```python
from mplsoccer import Pitch, VerticalPitch, Sbopen
from matplotlib.colors import LinearSegmentedColormap
parser = Sbopen()
```

```python
df_eventos, df2, df3, df4 = parser.event(3943077)
```

```python
pitch = Pitch(
    label=True,
    axis=True,
    pitch_type='statsbomb'
)

fig, ax = pitch.draw()
```

```python
df_eventos.player_name.value_counts()
```

```python
jugador = df_eventos[df_eventos['player_name'] == 'Jhon Adolfo Arias Andrade']
```

```python
jugador[['x', 'y']]
```

```python
pitch = Pitch(
    pitch_type='statsbomb',
    line_zorder=2,
    pitch_color='#f9fec5'
)

fig, ax = pitch.draw()

fig.set_facecolor('#f9fec5')

custom_cmap = LinearSegmentedColormap.from_list("Custom cmap",
                                                  ['#f9fec5', '#576000'],
                                                  N=100)

pitch.kdeplot(
    jugador.x,
    jugador.y,
    ax=ax,
    fill=True,
    levels=100,
    shade_lowest=True,
    zorder=-1,
    cmap=custom_cmap
)
```
