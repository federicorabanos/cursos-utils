```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Simulacin de datos
np.random.seed(42)
num_jugadores = 164

x = np.random.uniform(1, 6, num_jugadores)  # Duelos areos ganados por 90 min
y = np.random.uniform(40, 85, num_jugadores)  # % de duelos areos ganados
nombres = [f"Jugador {i}" for i in range(num_jugadores)]  # Nombres ficticios

# Seleccionar algunos jugadores aleatorios para etiquetar
jugadores_destacados = random.sample(range(num_jugadores), 15)

# Estilo del grfico
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# Scatter plot
plt.scatter(x, y, c="brown", alpha=0.7, edgecolors="black", linewidth=0.5)

# Etiquetar jugadores destacados
for i in jugadores_destacados:
    plt.text(x[i], y[i], nombres[i], fontsize=9, ha="left", va="bottom", fontweight="bold")

# Lneas de referencia en los promedios
plt.axhline(np.mean(y), color="teal", linestyle="--", alpha=0.7)
plt.axvline(np.mean(x), color="teal", linestyle="--", alpha=0.7)

# Ttulos y etiquetas
plt.title(" Estadsticas de defensores", fontsize=14, fontweight="bold", loc="left")
plt.suptitle("Duelos areos ganados por 90min vs. % de duelos areos ganados", fontsize=10, y=0.95)
plt.xlabel("Duelos areos ganados por 90min", fontsize=11)
plt.ylabel("% de Duelos areos ganados", fontsize=11)

# Nota inferior
plt.figtext(0.5, -0.05, "Jugadores con ms de 500 min. Cantidad de jugadores mostrados en el grfico: 164",
            wrap=True, horizontalalignment='center', fontsize=9)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parmetros personalizables
num_jugadores = 164
min_duelos_90 = 1
max_duelos_90 = 6
min_porcentaje = 40
max_porcentaje = 85
top_n = 15  # Cantidad de jugadores destacados

# Simulacin de datos
np.random.seed(42)
x = np.random.uniform(min_duelos_90, max_duelos_90, num_jugadores)  # Duelos areos ganados por 90 min
y = np.random.uniform(min_porcentaje, max_porcentaje, num_jugadores)  # % de duelos areos ganados
nombres = [f"Jugador {i}" for i in range(num_jugadores)]  # Nombres ficticios

# Seleccionar jugadores destacados en base a un ranking combinado
ranking = x / max_duelos_90 + y / max_porcentaje  # Combina ambos ejes
top_indices = np.argsort(ranking)[-top_n:]  # Selecciona los mejores

# Ajustar degradado basado en X e Y
color_intensidad = (x / max_duelos_90 + y / max_porcentaje) / 2  # Normaliza en base a los valores ms altos

# Estilo del grfico
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(7, 9))  # Ms alto que ancho, ideal para Twitter/X

# Scatter plot con degradado ajustado
ax.scatter(x, y, c=color_intensidad, cmap="Reds", alpha=0.8, edgecolors="black", linewidth=0.5)

# Etiquetar jugadores destacados
for i in top_indices:
    ax.text(x[i], y[i], nombres[i], fontsize=9, ha="left", va="bottom", fontweight="bold")

# Lneas de referencia (promedios)
ax.axhline(np.mean(y), color="teal", linestyle="--", alpha=0.6)
ax.axvline(np.mean(x), color="teal", linestyle="--", alpha=0.6)

# Ttulos y subttulos (AHORA FUERA DEL GRFICO)
plt.subplots_adjust(top=0.82)  # Ajusta espacio para los ttulos

fig.suptitle(" Estadsticas de defensores", fontsize=14, fontweight="bold", ha="left", x=0.12, fontfamily="sans-serif")
fig.text(0.12, 0.94, "Duelos areos ganados por 90min vs. % de duelos areos ganados", fontsize=10, fontfamily="sans-serif")
fig.text(0.12, 0.91, "Solo jugadores con ms de 1 y ms de 30, en cada estadstica respectivamente.", fontsize=9, color="gray", fontfamily="sans-serif")
fig.text(0.12, 0.88, "Visualizacin by @TuCuenta | Data: Fbref | Liga Profesional Argentina 2024", fontsize=9, color="gray", fontfamily="sans-serif")

# Etiquetas de los ejes
ax.set_xlabel("Duelos areos ganados por 90min", fontsize=11, fontfamily="sans-serif")
ax.set_ylabel("% de Duelos areos ganados", fontsize=11, fontfamily="sans-serif")

# Pie de foto alineado a la izquierda
fig.text(0.12, 0.03, f"Jugadores con ms de 500 min. Cantidad de jugadores mostrados en el grfico: {num_jugadores}", fontsize=9, fontfamily="sans-serif")

plt.show()
```
