```python
import requests
from bs4 import BeautifulSoup
```

```python
data = requests.get('https://www.transfermarkt.com.ar/kylian-mbappe/elfmetertore/spieler/342229/saison_id//wettbewerb_id//plus/1#tore', headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate',
})
```

```python
soup = BeautifulSoup(data.content, 'html.parser')
soup
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# Suponiendo que el HTML está en una variable `html`

# Encuentra la tabla
table = soup.find('table', {'class': 'items'})

# Encuentra todos los encabezados originales
all_th = table.find('thead').find_all('th')

# Crea una lista con los índices y nombres de columnas útiles
valid_indices = []
headers = []
for idx, th in enumerate(all_th):
    th_id = th.get('id', '')
    text = th.get_text(strip=True)
    if 'wappen' not in th_id and 'hide' not in th.get('class', []) and text:
        valid_indices.append(idx)
        headers.append(text)

# Extrae las filas
rows = []
for tr in table.find('tbody').find_all('tr'):
    tds = tr.find_all(['td', 'th'])
    row = []
    for idx in valid_indices:
        if idx < len(tds):
            text = tds[idx].get_text(strip=True)
            row.append(text)
    rows.append(row)

# Construir el DataFrame
df = pd.DataFrame(rows, columns=headers)

# Mostrar
print(df)
```
