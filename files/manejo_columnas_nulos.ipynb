```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Primera Division Argentina - passing.csv')
```

```python
df_pases = df[['Player', 'passing_Squad', 'passing_Pos', 'passing_90s', 'passing_Cmp']]
```

```python
df_pases['passing_Cmp'] = df_pases['passing_Cmp'].replace(0, np.nan)
df_pases
```

```python
df_pases.isna().sum()
```

```python
df_pases['passing_Cmp'] = df_pases['passing_Cmp'].fillna(0)
```

```python
df_pases['passing_Cmp_90s'] = df_pases['passing_Cmp'] / df_pases['passing_90s']
df_pases
```

```python
df_pases.sort_values(by=['passing_Cmp_90s', 'Player'], ascending=[True, False]).head(16)
```
