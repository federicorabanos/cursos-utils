```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Primera Division Argentina - passing.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
delanteros = df[(df['passing_Pos'] == 'FW')].reset_index(drop=True)
```

```python
delanteros.groupby('passing_Squad').sum()['passing_Ast'].sort_values(ascending=False)
```

```python
delanteros_agrupados = delanteros.groupby('passing_Squad')
```

```python
delanteros_agrupados.agg({
    'passing_Ast': ['sum', 'count'],
    'passing_Born': ['mean'],
    'passing_90s': ['median', 'std']
})
```
